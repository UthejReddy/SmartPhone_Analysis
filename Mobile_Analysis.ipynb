```python
mobile['EDGE']= mobile['EDGE'].map({'Yes':1, 'No':0})
mobile['RAM']= mobile['RAM'].map({'1 GB RAM':1, '2 GB RAM':2,'3 GB RAM':3})
mobile['audio_jack']= mobile['audio_jack'].map({'Yes':1, 'No':0})
mobile['radio']= mobile['radio'].map({'FM radio':1,'Stereo FM radio':2,'Stereo FM radio| RDS':3,'To be confirmed':4,'FM radio| RDS':5, 'No':0})
mobile['internal_memory']= mobile['internal_memory'].map({'4 GB':4,'8 GB':8,'16 GB':16,'32 GB':32,'512 GB':512})
mobile['status']= mobile['status'].map({'Available':1,'Discontinued':0,'Available. Released 2015  August':1,'Available. Released 2014  May ':1,'Available. Released 2013  October':1})
mobile['OS']= mobile['OS'].map({'Android 4.4.2 (KitKat)':4.4,'Android 2.2 (Froyo)':2.2,'Android 5.1 (Lollipop)':5.1,'Android 4.2 (Jelly Bean)':4.2,'Microsoft Windows Mobile 6.0 Professional':6.0})
mobile['loud_speaker']= mobile['loud_speaker'].map({'Yes':1,'Yes with stereo speakers':1, 'No':0})
```

```python
x = mobile_final.drop("approx_price_EUR", axis=1)  #independent columns
y = mobile_final["approx_price_EUR"]    #target column i.e PRICE
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import f_classif
```

```python
#apply SelectKBest class to extract top 17 best features
bestfeatures = SelectKBest(score_func=f_classif , k=8)
fit = bestfeatures.fit(x,y)
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
```

```python
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Features','Score']  #naming the dataframe columns
```

```python
print(featureScores.nlargest(8,"Score"))
```

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
feature = model.fit(x,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(9).plot(kind='barh')
plt.show()
```

# Removing Outliers

```python
lower_bound=0.1
upper_bound=0.95
res=mobile_final_9.approx_price_EUR.quantile([lower_bound,upper_bound])
res
```

```python
true_index=(res.loc[lower_bound]<mobile_final_9.approx_price_EUR.values)&\
           (mobile_final_9.approx_price_EUR.values<res.loc[upper_bound])
true_index
```

```python
false_index=~true_index
false_index
```

```python
mobile_final_9.approx_price_EUR[true_index]
```

```python
mid=np.median(mobile_final_9.approx_price_EUR[true_index])
mid
```

```python
mobile_final_9[false_index]=mid
```
